```python
# stack  간단 구현 
# 스택 : FILO,  LIFO  ->   입구와 출구가 하나
stack = []
for _ in range(5):    # stack 리스트에 5개의 None을 추가
    stack.append(None)
print(stack)

# stack의 자료의 위치 -> top = -1 초기화
top = -1

#  stack에 '커피'를 추가
if top >= len(stack) - 1 :
    print("더 이상 추가할 공간 없음 ")
else:
    top += 1
    stack[top] = '커피'
    
#  stack에 '꿀물'을 추가
if top >= len(stack) - 1 :
    print("더 이상 추가할 공간 없음 ")
else:
    top += 1
    stack[top] = '꿀물'    
    
#  stack의 전체 자료 출력
for i in range(len(stack)-1, -1, -1):
    print(stack[i])
print()

# stack에 저장된 자료만 출력
for i in range(top, -1, -1):
    print(stack[i])
```

```python
def isStackFull():
    global SIZE, top, stack
    if top >= SIZE -1:
        return True
    else:
        return False

def isStackEmpty():
    global SIZE, top, stack
    if top == -1:
        return True
    else:
        return False
    
def stack_push(data):
    global SIZE, top, stack
    if isStackFull():
        print("더 이상 추가할 공간 없음 ")
        return
    top += 1
    stack[top] = data 
    
def stack_pop():
    global SIZE, top, stack
    if isStackEmpty():
        print("자료 없음")
        return 
    data = stack[top]
    stack[top] = None
    top -= 1
    return data
```

```python
# stack size
SIZE = 5
stack = [ None for _ in range(SIZE) ]  # 리스트 컴프리핸션을 사용
top = -1

for i in range(SIZE):
    stack_push(str(i))
```

```python
# 문 : 스택의 사이즈 입력
# 1. 입력된 사이즈 만큼 리스트 생성 -> 초기값은 None
# 2.  메뉴입력 (1 : 삽입, 2: 추출, 3: 리스트 4: 종료) -> 
# 3. push(), pop(), list()  함수로 구성
#  추가의 함수는 push할 때 스택이 다 찼는지 확인
#                pop 또는 list 를 실행할 경우 스택이 비어 있는지 확인
# isfull(), isempty()
# 1. 
def isfull():
    global size, stack, top
    if top >= size -1 :
        return True
    else:
        return False

def isempty():
    global size, stack, top
    if top == -1 :
        return True
    else:
        return False
    
def push(data):
    global size, stack, top
    #  스택이  Full 인지 확인
    if isfull():
        print('입력할 공간 없음')
        return
    top += 1
    stack[top] = data
    
def pop():
    global size, stack, top
    #  스택이  empty 인지 확인
    if isempty():
        print('추출할 자료 없음')
        return
    data = stack[top]
    stack[top] = None
    top -= 1
    return data

def peek():
    global size, stack, top
    #  스택이  empty 인지 확인
    if isempty():
        print(' 자료 없음')
        return    
    return stack[top]

def stack_list():
    global size, stack, top
    #  스택이  empty 인지 확인
    if isempty():
        print(' 저장된 자료 없음')
        return
    
    for i in range(top, -1, -1):
        print(stack[i], end='\t')
    print()
    
size = int(input("스택 사이즈 입력 : "))
stack = [ None for _ in range(size)]
top = -1   # 스택의 마지막 자료 위치
# 2. 
while True:
    menu = input("메뉴입력 (1 : 삽입, 2: 추출, 3:마지막 자료확인 4: 리스트 5: 종료) -> ")
    if menu == '1':
        data = input(' 삽입할 자료 입력 : ')
        push(data)
    elif menu == '2':
        data = pop()
        if data:
            print("추출한 자료는 : ", data)
    elif menu == '3':
        data = peek()    
        if data:
            print("마지막 저장 자료는 : ", data)        
    elif menu == '4':
        stack_list()
    elif menu == '5':
        break
```

```python
def isFull():
    global size, stack, top
    if top >= size -1 :
        return True
    else:
        return False

def isEmpty():
    global top
    if top == -1 :
        return True
    else:
        return False
    
def Push(data):
    global size, stack, top
    #  스택이  Full 인지 확인
    if isFull():
        print('입력할 공간 없음')
        return
    top += 1
    stack[top] = data
    return stack
    
def pop():
    global size, stack, top
    #  스택이  empty 인지 확인
    if isempty():
        print('추출할 자료 없음')
        return
    data = stack[top]
    stack[top] = None
    top -= 1
    return data

def peek():
    global size, stack, top
    #  스택이  empty 인지 확인
    if isempty():
        print(' 자료 없음')
        return    
    return stack[top]

def stack_list():
    global size, stack, top
    #  스택이  empty 인지 확인
    if isempty():
        print(' 저장된 자료 없음')
        return
    
    for i in range(top, -1, -1):
        print(stack[i], end='\t')
    print()
```

```python
import webbrowser
import time

size = 100
stack = [ None for _ in range(size)]
top = -1   # 스택의 마지막 자료 위치

urls = ['naver.com', 'nate.com', 'daum.net']
# urls의 각 요소를  stack에  저장
for url in urls:
    push(url)
    webbrowser.open('http://' + url)
    print(url , end=' -> ')
    time.sleep(1)
print("방문 종료")
time.sleep(5)

while True:
    url = pop()
    if url == None:
        break
    webbrowser.open('http://' + url)
    print(url , end=' -> ')
    time.sleep(1)
print('방문 종료')
```

```python
def checkBracket(expr):
    for char in expr:
        if char in '([<{':
            push(char)
        elif char in ')]>}':
            if top == -1:
                return False
            data = pop()
            if char == ')' and data == '(':
                pass
            elif char == ']' and data == '[':
                pass
            elif char == '>' and data == '<':
                pass
            elif char == '}' and data == '{':
                pass 
            else:
                return False
    if isempty():
        return True
    else:
        return False
```

```python
# 수식을 쌍으로 맞는지 확인 -> 스택을 활용하여 check
size = 100
stack = [ None for _ in range(size)]

exprarray = [ '(A+B)', ')A+B(', '((A+B) -C', '(A+B]', '(<a+{b-c}/[c*d]>)']

for expr in exprarray:
    top = -1
#     print('... ', expr)
    print(expr, ' ==> ', checkBracket(expr))
```
