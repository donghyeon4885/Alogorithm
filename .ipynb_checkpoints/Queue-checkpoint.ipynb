```python
## Queue : 출구와 입구가 다른 자료구조 ->  선형 큐
def isQueuefull():
    global size, queue, front, rear
    if rear == size-1:
        return True
    else:
        return False
    
def isQueueempty():
    global size, queue, front, rear
    if front == rear:
        return True
    else:
        return False    
        
def enQueue(data):
    global size, queue, front, rear
    if isQueuefull():
        print(" 큐가 다 찼음")
        return
    rear += 1
    queue[rear] = data

def deQueue():
    global size, queue, front, rear
    if isQueueempty():
        print(" 큐가 비었음")
        return
    front += 1  
    data = queue[front]
    queue[front] = None
    
    #  뒤에서 앞으로 데이터 이동
    for i in range(front+1, rear + 1):
        queue[i - 1] = queue[i]
        queue[i] = None
    front = -1   # 출구의 위치를 초기값으로 지정
    rear -= 1    # 입구의 위치 1칸 앞으로 이동
    return data

def peekQueue():  # 추출될 데이터 확인
    global size, queue, front, rear
    if isQueueempty():
        print(" 큐가 비었음")
        return
    return queue[front + 1]

def printQueue():
    global size, queue, front, rear    
    print(" ---  큐의 상태 ----")
    print(" 출구 <-- ", end=' ')
    for i in range(len(queue)):
        print(queue[i], end=' ')
    print(" --> 입구 ")
```

```python
size = int(input("큐 사이즈 입력 : "))
queue = [ None for _ in range(size)]
front = rear = -1  #  출구(front),  입구(rear) 를 초기화
# 2. 
while True:
    menu = input("메뉴입력 (1 : 삽입, 2: 추출, 3:추출할 자료 확인 4: 리스트 5: 종료) -> ")
    if menu == '1':
        data = input(' 삽입할 자료 입력 : ')
        enQueue(data)
    elif menu == '2':
        data = deQueue()
        if data:
            print("추출한 자료는 : ", data)      
    elif menu == '3':
        data = peekQueue()     
        if data:
            print("추출할 자료는 : ", data)         
    elif menu == '4':
        printQueue()
    elif menu == '5':
        break
```

```python
## Queue : 출구와 입구가 다른 자료구조 ->  원형 큐
def isfull():
    global size, queue, front, rear
    if (rear + 1) % size == front:    # 자료가  Full
        return True
    else:
        return False
    
def isempty():
    global size, queue, front, rear
    if front == rear:
        return True
    else:
        return False    
        
def enQueue(data):
    global size, queue, front, rear
    if isfull():
        print(" 큐가 다 찼음")
        return
    rear = (rear + 1) % size
    queue[rear] = data

def deQueue():
    global size, queue, front, rear
    if isempty():
        print(" 큐가 비었음")
        return
    front = (front + 1) % size
    data = queue[front]
    queue[front] = None
    
    return data

def peekQueue():  # 추출될 데이터 확인
    global size, queue, front, rear
    if isempty():
        print(" 큐가 비었음")
        return
    return queue[ (front + 1) % size ]

def printQueue():
    global size, queue, front, rear    
    print(" ---  큐의 상태 ----")
    print(" 출구 <-- ", end=' ')
    for i in range(size):
        print(queue[i], end=' ')
    print(" --> 입구 ")
    print(f'front == {front},  rear == {rear}')
```

```python
size = int(input("큐 사이즈 입력 : "))
queue = [ None for _ in range(size)]
front = rear = 0  #  출구(front),  입구(rear) 를 초기화,  원형 큐의 경우 0부터 시작

while True:
    menu = input("메뉴입력 (1 : 삽입, 2: 추출, 3:추출할 자료 확인 4: 리스트 5: 종료) -> ")
    if menu == '1':
        data = input(' 삽입할 자료 입력 : ')
        enQueue(data)
    elif menu == '2':
        data = deQueue()
        if data:
            print("추출한 자료는 : ", data)      
    elif menu == '3':
        data = peekQueue()     
        if data:
            print("추출할 자료는 : ", data)         
    elif menu == '4':
        printQueue()
    elif menu == '5':
        break
    
```
