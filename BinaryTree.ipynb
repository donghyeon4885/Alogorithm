```python
# 이진 트리 구조 실습
class TreeNode():
    def __init__(self):
        self.left = None
        self.data = None
        self.right = None
```

```python
node1 = TreeNode()  #  빈 노드 생성
node1.data = '화사'

node2 = TreeNode()
node2.data = '솔라'
# node1의 left 에 node2  연결
node1.left = node2

node3 = TreeNode()
node3.data = '문별'
# node1의 right 에 node3  연결
node1.right = node3

node4 = TreeNode()
node4.data = '휘인'
# node2의 left 에 node4  연결
node2.left = node4

node5 = TreeNode()
node5.data = '쯔위'
# node2의 rigth 에 node5  연결
node2.right = node5

node6 = TreeNode()
node6.data = '선미'
# node3의 left 에 node6  연결
node3.left = node6
```

```python
#  함수로 정의 : 3가지 방법으로
# 전위 함수 : 현재 데이터 처리 -> 왼쪽 -> 오른쪽
def preorder(node):
    if node == None:
        return
    print(node.data, end= ' -> ' )
    preorder(node.left)
    preorder(node.right)
    
# 중위 함수 : 현재  왼쪽 -> 데이터 처리 ->오른쪽
def inorder(node):
    if node == None:
        return
    inorder(node.left)    
    print(node.data, end= ' -> ' )
    inorder(node.right)    
    
# 후위 함수 : 현재  왼쪽 ->오른쪽 ->데이터 처리
def postorder(node):
    if node == None:
        return
    postorder(node.left)  
    postorder(node.right)    
    print(node.data, end= ' -> ' )
        
```

```python
print("전위 순회 -> ", end=' : ')
preorder(node1)
print("end")

print("중위 순회 -> ", end=' : ')
inorder(node1)
print("end")

print("후위 순회 -> ", end=' : ')
postorder(node1)
print("end")
```

```python
#  순차 이진 트리 완성
class TreeNode():
    def __init__(self):
        self.left = None
        self.data = None
        self.right = None

names = ['블랙핑크', '레드벨벳', '마마무','에이핑크' ,'걸스데이', '트와이스']
root = None
node = TreeNode()
node.data = names[0] # 블랙핑크를  root로 지정

root = node
for name in names[1:]:
    node = TreeNode()
    node.data = name
    
    #  이진 트리 추가
    curr_node = root
    while True:
        if name < curr_node.data:  #  left node 로 추가
            if curr_node.left == None:
                curr_node.left = node
                break
            curr_node = curr_node.left  # 다음 검색할 node로 이동
        else:    # right node로 추가
            if curr_node.right == None:
                curr_node.right = node
                break
            curr_node = curr_node.right  # 다음 검색할 node로 이동
    
print(" 이진 검색 트리 완성 ", preorder(root))
```

```python
#  순차 이진 트리 완성 -->  insert_tree(new_node)
class TreeNode():
    def __init__(self):
        self.left = None
        self.data = None
        self.right = None

def insert_node(new_node):
    global root
    
    if root == None:   # 처음 자료이면  root로 지정
        root = new_node
        return
    
    node = root   #  처음 시작 위치 
    while True:
        if new_node.data < node.data:  #  left node 로 추가
            if node.left == None:
                node.left = new_node
                return
            node = node.left  # 다음 검색할 node로 이동
        else:    # right node로 추가
            if node.right == None:
                node.right = new_node
                return
            node = node.right  # 다음 검색할 node로 이동
```

```python
root = None
while True:
    input_data = input('입력할 자료 입력 : ')
    if not input_data:
        break
    node = TreeNode()
    node.data = input_data
    insert_node(node)
    inorder(root)
    print()
print('end')
```
